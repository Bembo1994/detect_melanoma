```python
import os, random, cv2
import numpy as np
from datetime import datetime
from keras.models import load_model
from tensorflow.keras.utils import normalize
from matplotlib import pyplot as plt

from tensorflow.keras import backend as K
from tf_keras_vis.saliency import Saliency
from tf_keras_vis.utils.scores import BinaryScore
from tf_keras_vis.utils.model_modifiers import ReplaceToLinear
from matplotlib import cm
from tf_keras_vis.activation_maximization import ActivationMaximization
from tf_keras_vis.activation_maximization.callbacks import Progress
from tf_keras_vis.activation_maximization.input_modifiers import Jitter, Rotate2D, Scale
from tf_keras_vis.activation_maximization.regularizers import Norm, TotalVariation2D
import tensorflow as tf


def read_in_rgb(path):
    imm = cv2.imread(path)
    return cv2.cvtColor(imm, cv2.COLOR_BGR2RGB)

def mole_detect(origin_image_rgb, mask, delta):

    contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    contour = None
    max_area = 0
    for c in contours:
        area = cv2.contourArea(c)
        if area > max_area:
            max_area = area
            contour = c
    boundRect = cv2.boundingRect(contour)

    output = cv2.bitwise_and(origin_image_rgb, origin_image_rgb, mask=mask)
    r = cv2.drawContours(image=output, contours=contour, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
        
    x1, y1, w, h = boundRect
    
    x2 = x1 + w + delta
    y2 = y1 + h + delta
    
    if x1 - delta <= 0 : 
        x1 = 0
    else :
        x1 = x1 - delta

    if x2 >= origin_image_rgb.shape[1]:
        x2 = origin_image_rgb.shape[1]# - w - delta
    
    if y1 - delta <= 0 : 
        y1 = 0
    else :
        y1 = y1 - delta

    if y2 >= origin_image_rgb.shape[0]:
        y2 = origin_image_rgb.shape[0]# - h - delta
    
    bby = y2 - y1
    bbx = x2 - x1

    
    if bbx/bby < 1:
        diff = bby - bbx
        while diff > 0 :
            count_tmp = 0
            if x1 - 1 > 0 and diff - (count_tmp + 1) >= 0:
                x1 -= 1
                count_tmp += 1
            if x2 + 1 <= origin_image_rgb.shape[1] and diff - (count_tmp + 1) >= 0:
                x2 += 1
                count_tmp += 1
                
            if count_tmp > 1:
                diff -= count_tmp
            else:
                diff -= 1          
    bby = y2 - y1
    bbx = x2 - x1
    
    if bbx/bby > 1:
        diff = bbx - bby
        while diff > 0 :
            count_tmp = 0
            if y1 - 1 > 0 and diff - (count_tmp + 1) >= 0:
                y1 -= 1
                count_tmp += 1
            if y2 + 1 <= origin_image_rgb.shape[0] and diff - (count_tmp + 1) >= 0:
                y2 += 1
                count_tmp += 1
                
            if count_tmp > 1:
                diff -= count_tmp
            else:
                diff -= 1
    bby = y2 - y1
    bbx = x2 - x1
    
    rectangle = cv2.rectangle(r, (x1, y1), (x2, y2), (0, 255, 0), 2)

    f = origin_image_rgb[y1:y2, x1:x2]
    k = cv2.resize(f, (299, 299), interpolation=cv2.INTER_CUBIC)
    k = k/255.
    
    return mask, rectangle, k

def unet_preprocessing(path, model, delta):
    start=datetime.now()
    original = read_in_rgb(path)

    original_resize = cv2.resize(original, (256, 256), interpolation=cv2.INTER_CUBIC)
    img_to_gray = cv2.cvtColor(original_resize, cv2.COLOR_RGB2GRAY)
    img_norm = np.expand_dims(normalize(np.array(img_to_gray), axis=1), 2)
    img_norm = img_norm[:, :, 0][:, :, None]
    to_input = np.expand_dims(img_norm, 0)
    # Predict and threshold for values above 0.5 probability
    # Change the probability threshold to low value (e.g. 0.05) for watershed demo.
    prediction_mask = (model.predict(to_input)[0, :, :, 0] > 0.2).astype(np.uint8)
    
    pred_mask = cv2.resize(prediction_mask, (original.shape[1], original.shape[0]), interpolation=cv2.INTER_CUBIC)

    m, r, k = mole_detect(original, pred_mask, delta)
    #print("Time for preprocessing with UNET : {}".format(datetime.now()-start))
    return m, r, k

def cv_preprocessing(path, delta):
    start=datetime.now()
    original = read_in_rgb(path)
   
    # Convert to HSV colourspace and extract just the Saturation
    saturation = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)[..., 1]
    # Find best (Otsu) threshold to divide black from white, and apply it
    _, mask = cv2.threshold(saturation, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #cv2.threshold(saturation, 0.5, 1, 0)
    m, r, k = mole_detect(original, mask, delta)
    #print("Time for preprocessing with CV : {}".format(datetime.now()-start))
    return  m, r, k

path_benign_train_ds = "dataset/ISICArchive/Classification_totali/train/benign/"
#path_malignant_train_ds = "/home/bembo/Scrivania/detect_melanoma/dataset/ISICArchive/Classification/malignant/"


unet_model = load_model("machine_learning/checkpoints/unet_relu_None_lr_0-001_bs_20_rmsprop")
def take_random_img(numb, path):
    
    images = os.listdir(path)
    list_to_return = set()
    
    while len(list_to_return) < numb:
        r = random.choice(images)
        list_to_return.add(r)
        
    return list_to_return


name_imgs_benign = take_random_img(50, path_benign_train_ds)
#'ISIC_0012406.jpg'
#'ISIC_0011090.jpg', 'ISIC_0026279.jpg', 'ISIC_0015412.jpg', 'ISIC_0010108.jpg'
name_imgs_benign = ['ISIC_8246244.jpg']
#name_imgs_benign = ['ISIC_0011090.jpg', 'ISIC_0026279.jpg', 'ISIC_0015412.jpg', 'ISIC_0010108.jpg', 'ISIC_0010691.jpg']#['ISIC_0024827.jpg', 'ISIC_0000095.jpg', 'ISIC_0010788.jpg', 'ISIC_0011108.jpg', 'ISIC_0012406.jpg', 'ISIC_0012209.jpg']
for img in name_imgs_benign:
    
    print(img)
    cv_delta = 30
    cv_mask, cv_bounding_box, cv_result = cv_preprocessing(path_benign_train_ds+img, cv_delta)

    plt.figure(figsize=(12, 4))
    plt.subplot(151)
    plt.title('Originale')
    plt.imshow(read_in_rgb(path_benign_train_ds+img))
    plt.xticks([])
    plt.yticks([])
    plt.subplot(152)
    plt.title('Maschera Threshold')
    plt.imshow(cv_mask)
    plt.xticks([])
    plt.yticks([])
    plt.subplot(153)
    plt.title('Threshold Bounding Box')
    plt.imshow(cv_bounding_box)
    plt.xticks([])
    plt.yticks([])
    plt.subplot(154)
    plt.title('Threshold Risultato')
    plt.imshow(cv_result)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("img_for_thesis/Threshold_processing_Confronto_2.png")
    plt.show()
    
    
    unet_delta = 30
    unet_mask, unet_bounding_box, unet_result = unet_preprocessing(path_benign_train_ds+img, unet_model, unet_delta)

    plt.figure(figsize=(12, 4))
    plt.subplot(151)
    plt.title('Originale')
    plt.imshow(read_in_rgb(path_benign_train_ds+img))
    plt.xticks([])
    plt.yticks([])
    plt.subplot(152)
    plt.title('Maschera con UNET')
    plt.imshow(unet_mask)
    plt.xticks([])
    plt.yticks([])
    plt.subplot(153)
    plt.title('UNET Bounding Box')
    plt.imshow(unet_bounding_box)
    plt.xticks([])
    plt.yticks([])
    plt.subplot(154)
    plt.title('UNET Risultato')
    plt.imshow(unet_result)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("img_for_thesis/UNET_processing_Confronto_2.png")
    plt.show()
#{'ISIC_0024827.jpg'}
```

```python
import os, random, cv2
import numpy as np
from datetime import datetime
from keras.models import load_model
from tensorflow.keras.utils import normalize
from matplotlib import pyplot as plt

from tensorflow.keras import backend as K
from tf_keras_vis.saliency import Saliency
from tf_keras_vis.utils.scores import BinaryScore
from tf_keras_vis.utils.model_modifiers import ReplaceToLinear
from matplotlib import cm
from tf_keras_vis.activation_maximization import ActivationMaximization
from tf_keras_vis.activation_maximization.callbacks import Progress
from tf_keras_vis.activation_maximization.input_modifiers import Jitter, Rotate2D, Scale
from tf_keras_vis.activation_maximization.regularizers import Norm, TotalVariation2D
import tensorflow as tf


def read_in_rgb(path):
    imm = cv2.imread(path)
    return cv2.cvtColor(imm, cv2.COLOR_BGR2RGB)

def mole_detect(origin_image_rgb, mask, delta):

    contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
    contour = None
    max_area = 0
    for c in contours:
        area = cv2.contourArea(c)
        if area > max_area:
            max_area = area
            contour = c
    boundRect = cv2.boundingRect(contour)

    output = cv2.bitwise_and(origin_image_rgb, origin_image_rgb, mask=mask)
    r = cv2.drawContours(image=output, contours=contour, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
   
    
    x1, y1, w, h = boundRect
    

    
    x2 = x1 + w + delta
    y2 = y1 + h + delta
    

    if x1 - delta <= 0 : 
        x1 = 0
    else :
        x1 = x1 - delta

    if x2 >= origin_image_rgb.shape[1]:
        x2 = origin_image_rgb.shape[1]# - w - delta
    
    if y1 - delta <= 0 : 
        y1 = 0
    else :
        y1 = y1 - delta

    if y2 >= origin_image_rgb.shape[0]:
        y2 = origin_image_rgb.shape[0]# - h - delta
    
    bby = y2 - y1
    bbx = x2 - x1

    
    if bbx/bby < 1:
        diff = bby - bbx
        while diff > 0 :
            count_tmp = 0
            if x1 - 1 > 0 and diff - (count_tmp + 1) >= 0:
                x1 -= 1
                count_tmp += 1
            if x2 + 1 <= origin_image_rgb.shape[1] and diff - (count_tmp + 1) >= 0:
                x2 += 1
                count_tmp += 1
                
            if count_tmp > 1:
                diff -= count_tmp
            else:
                diff -= 1          
    bby = y2 - y1
    bbx = x2 - x1
    
    if bbx/bby > 1:
        diff = bbx - bby
        while diff > 0 :
            count_tmp = 0
            if y1 - 1 > 0 and diff - (count_tmp + 1) >= 0:
                y1 -= 1
                count_tmp += 1
            if y2 + 1 <= origin_image_rgb.shape[0] and diff - (count_tmp + 1) >= 0:
                y2 += 1
                count_tmp += 1
                
            if count_tmp > 1:
                diff -= count_tmp
            else:
                diff -= 1
    bby = y2 - y1
    bbx = x2 - x1
    
    rectangle = cv2.rectangle(r, (x1, y1), (x2, y2), (0, 255, 0), 2)
   
    return rectangle

def unet_preprocessing(path, model, delta):
    start=datetime.now()
    original = read_in_rgb(path)

    original_resize = cv2.resize(original, (256, 256), interpolation=cv2.INTER_CUBIC)
    img_to_gray = cv2.cvtColor(original_resize, cv2.COLOR_RGB2GRAY)
    img_norm = np.expand_dims(normalize(np.array(img_to_gray), axis=1), 2)
    img_norm = img_norm[:, :, 0][:, :, None]
    to_input = np.expand_dims(img_norm, 0)
    # Predict and threshold for values above 0.5 probability
    # Change the probability threshold to low value (e.g. 0.05) for watershed demo.
    prediction_mask = (model.predict(to_input)[0, :, :, 0] > 0.2).astype(np.uint8)
    
    pred_mask = cv2.resize(prediction_mask, (original.shape[1], original.shape[0]), interpolation=cv2.INTER_CUBIC)

    m, r, k = mole_detect(original, pred_mask, delta)
    #print("Time for preprocessing with UNET : {}".format(datetime.now()-start))
    return m, r, k

def cv_preprocessing(path, delta):
    start=datetime.now()
    original = read_in_rgb(path)
   
    # Convert to HSV colourspace and extract just the Saturation
    saturation = cv2.cvtColor(original, cv2.COLOR_RGB2HSV)[..., 1]
    # Find best (Otsu) threshold to divide black from white, and apply it
    _, mask = cv2.threshold(saturation, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #cv2.threshold(saturation, 0.5, 1, 0)
    o = mole_detect(original, mask, delta)
    #print("Time for preprocessing with CV : {}".format(datetime.now()-start))
    return  o

path_benign_train_ds = "dataset/ISICArchive/Classification_totali/train/benign/"
#path_malignant_train_ds = "/home/bembo/Scrivania/detect_melanoma/dataset/ISICArchive/Classification/malignant/"


unet_model = load_model("machine_learning/checkpoints/unet_relu_None_lr_0-001_bs_20_rmsprop")
def take_random_img(numb, path):
    
    images = os.listdir(path)
    list_to_return = set()
    
    while len(list_to_return) < numb:
        r = random.choice(images)
        list_to_return.add(r)
        
    return list_to_return


name_imgs_benign = take_random_img(50, path_benign_train_ds)
#'ISIC_0012406.jpg'
#'ISIC_0011090.jpg', 'ISIC_0026279.jpg', 'ISIC_0015412.jpg', 'ISIC_0010108.jpg'
name_imgs_benign = ['ISIC_0033298.jpg']
#name_imgs_benign = ['ISIC_0011090.jpg', 'ISIC_0026279.jpg', 'ISIC_0015412.jpg', 'ISIC_0010108.jpg', 'ISIC_0010691.jpg']#['ISIC_0024827.jpg', 'ISIC_0000095.jpg', 'ISIC_0010788.jpg', 'ISIC_0011108.jpg', 'ISIC_0012406.jpg', 'ISIC_0012209.jpg']
for img in name_imgs_benign:
    
    print(img)
    cv_delta = 30
    o = cv_preprocessing(path_benign_train_ds+img, cv_delta)

    plt.figure(figsize=(12, 4))
    plt.subplot(171)
    plt.title('Originale')
    plt.imshow(o)
    plt.xticks([])
    plt.yticks([])
    plt.savefig("img_for_thesis/output_5.png")
    plt.show()

    
```
